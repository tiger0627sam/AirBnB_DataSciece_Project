```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error ,r2_score, mean_absolute_error
from sklearn.model_selection import GridSearchCV
```

```python
df = pd.read_csv('/content/1126 Preprocessed Data.xlsx - Sheet1.csv')
voting_4_df = pd.read_csv('/content/1202 Voting_4_Feature.xlsx - Sheet1.csv')
voting_3_4_df = pd.read_csv('/content/1202 Voting_3&4_Feature.xlsx - Sheet1.csv')
elastic_net_df = pd.read_csv('/content/1202 Elastic_Net_Feature.xlsx - Sheet1.csv')
```

```python
voting_4 = voting_4_df['feature'].tolist()
voting_3_4 = voting_3_4_df['feature'].tolist()
elastic_net = elastic_net_df['feature'].tolist()
```

```python
features1 = df[voting_4]
features2 = df[voting_3_4]
features3 = df[elastic_net]
target = df['price']
```

**Voting 4**

```python
# voting 4

scaler = StandardScaler()
features_scaled = scaler.fit_transform(features1)

X_train, X_test, y_train, y_test = train_test_split(features_scaled, target, test_size=0.2, random_state=42)
n = len(y_test)
k = X_test.shape[1]
# poly模型评估
poly_svr = SVR(kernel='poly', C=3.0, degree = 3)
poly_svr.fit(X_train, y_train)
poly_y_pred = poly_svr.predict(X_test)

poly_r2 = r2_score(y_test, poly_y_pred)
poly_adjusted_r2 = 1 - (1 - poly_r2) * (n - 1) / (n - k - 1)
poly_rmse = mean_squared_error(y_test, poly_y_pred, squared=False)
poly_mae = mean_absolute_error(y_test, poly_y_pred)
print(f'Poly:Root Mean Squared Error: {poly_rmse}')
print(f'Linear:Adj-R² Score: {poly_adjusted_r2}')
print('----------------------------------------')

# linear模型評估
linear_svr = SVR(kernel='linear', C = 3.0, gamma = 0.01, epsilon = 0.01)
linear_svr.fit(X_train, y_train)
linear_y_pred = linear_svr.predict(X_test)

linear_rmse = mean_squared_error(y_test, linear_y_pred, squared=False)
linear_r2 = r2_score(y_test, linear_y_pred)
linear_adjusted_r2 = 1 - (1 - linear_r2) * (n - 1) / (n - k - 1)
linear_mae = mean_absolute_error(y_test, linear_y_pred)
print(f'Linear:Root Mean Squared Error: {linear_rmse}')
print(f'Linear:Adj-R² Score: {linear_adjusted_r2}')
print('----------------------------------------')

# rbf模型评估
rbf_svr = SVR(kernel='rbf', C=3.0, gamma = 0.01, epsilon = 0.01)
rbf_svr.fit(X_train, y_train)
rbf_y_pred = rbf_svr.predict(X_test)

rbf_rmse = mean_squared_error(y_test, rbf_y_pred, squared=False)
rbf_r2 = r2_score(y_test, rbf_y_pred)
rbf_mae = mean_absolute_error(y_test, rbf_y_pred)
rbf_adjusted_r2 = 1 - (1 - rbf_r2) * (n - 1) / (n - k - 1)
print(f'rbf:Root Mean Squared Error: {rbf_rmse}')
print(f'rbf:Adj-R² Score: {rbf_adjusted_r2}')
```

**K-fold Validation**

Voting 4

```python
from sklearn.model_selection import GridSearchCV, KFold
from sklearn.svm import SVR
from sklearn.metrics import make_scorer, mean_squared_error, r2_score, mean_absolute_error
import numpy as np

kernels = ['linear', 'poly', 'rbf']

for kernel in kernels:
    svr = SVR(kernel= kernel)
    param_grid = {
        'C': [0.1, 1, 3, 5],
        'gamma': [0.01, 0.1, 0.5, 1],
        'epsilon': [0.01, 0.1, 0,5, 1]
    }
    scaler = StandardScaler()
    features_scaled = scaler.fit_transform(features1)
    X_train, X_test, y_train, y_test = train_test_split(features_scaled, target, test_size=0.2, random_state=42)
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    scoring = {'neg_mean_absolute_error': 'neg_mean_absolute_error'}

    grid_search = GridSearchCV(estimator=svr, param_grid=param_grid, scoring=scoring, cv=kf, refit='neg_mean_absolute_error', return_train_score=True)
    grid_search.fit(X_train, y_train)
    best_params = grid_search.best_params_
    print(f'Best Parameters: {best_params}')

    best_svr_model = grid_search.best_estimator_
    y_pred = best_svr_model.predict(X_test)

    r2 = r2_score(y_test, y_pred)
    n = len(y_test)
    k = X_test.shape[1]
    adjusted_r2 = 1 - (1 - r2) * (n - 1) / (n - k - 1)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    print(f'Kernel:{kernel}')
    print(f'Adjusted R²: {adjusted_r2}')
    print(f'RMSE: {rmse}')
    print('--------------------------------------')
```

```python
import pandas as pd
user_field = pd.read_csv('/content/dgs_UserField.csv')
shop_order = pd.read_csv('/content/dgs_ShopOrder.csv', low_memory=False)
#user_field
#shop_order
```

```python
columns = ['D0', 'Id', 'Identity Code', 'Receipt Type', 'Send At', 'Updated At', 'userid', '下單日期', '公司', 'ODER天數', 'Order Status', 'Pay Cart Price Total', 'Pay Price Total', '前期訂單數', '當期同期成長率', '當期訂單數']
shop_order = shop_order[columns]
```

```python
def get_first_letter(company_id):
    if pd.notna(company_id):
        return company_id[0]
    else:
        return 'Z'

# 新增一个'group'列，并根据'Company Id'列的第一个字母填充
user_field['group'] = user_field['Company Id'].apply(get_first_letter)
```
